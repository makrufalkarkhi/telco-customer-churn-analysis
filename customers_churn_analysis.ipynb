### Import & Load Data

```python
import pandas as pd
df = pd.read_csv ('telco_customers_churn.csv')
```

```python
df.head()
```

### Data Overview

```python
df.info()
```

```python
df.describe(include='all')
```

### Missing Values

```python
df.isnull().sum()
```

### Fix Data Types & Missing Values

```python
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors='coerce')
```

```python
df[['Total Charges']].info()
```

```python
df['Total Charges'] = df['Total Charges'].fillna(df['Total Charges'].median())
```

```python
df['Churn Reason'] = df['Churn Reason'].fillna('Not Churned')
```

```python
df.isnull().sum()
```

### Normalize Column Names

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
```

```python
df.columns
```

### Feature Engineering

```python
df['tenure_group'] = pd.cut(
    df['tenure_months'],
    bins=[0, 12, 24, 48, 60, 100],
    labels=['0-1y', '1-2y', '2-4y', '4-5y', '5y+']
)
```

```python
df['is_high_value'] = df['cltv'] > df['cltv'].quantile(0.75)
```

```python
df['is_auto_pay'] = df['payment_method'].str.contains('automatic')
```

```python
df[['tenure_months','tenure_group','cltv','is_high_value','payment_method','is_auto_pay']].head(10)
```

### Check Redundant Columns

```python
(df['churn_label'] == (df['churn_value'] == 1).map({True:'Yes', False:'No'})).all()
```

```python
df = df.drop(columns=['churn_label'])
```

### Final Check

```python
df.to_csv("telco_customers_churn_cleaned.csv", index=False)
```
